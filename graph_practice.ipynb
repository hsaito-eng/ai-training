```python
%reset
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x, w):
    return (x - w) * x * (x + 2)

x = np.linspace(-3, 3, 100)
plt.plot(x, f(x, 2), color = 'black', label = '$w = 2$')
plt.plot(x, f(x, 1), color = 'cornflowerblue', label = '$w = 1$')
plt.legend(loc = 'upper left')
plt.ylim(-15, 15)
plt.title('$f(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

```python
%reset
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x0, x1):
    ans = (2 * x0**2 + x1**2) * np.exp(-(2 * x0**2 + x1**2))
    return ans

xn = 9

x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f(x0[i0], x1[i1])


plt.figure(figsize =(3.5, 3))
plt.gray()
plt.pcolor(y)
plt.colorbar()
plt.show()
```
